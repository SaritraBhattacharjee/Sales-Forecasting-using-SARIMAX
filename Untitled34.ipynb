```python
### Testing For Stationarity

from statsmodels.tsa.stattools import adfuller
```

```python
test_result=adfuller(df['Sales'])
```

```python
#Ho: It is non stationary
#H1: It is stationary

def adfuller_test(sales):
    result=adfuller(sales)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
        
```

```python
adfuller_test(df['Sales'])
```

```python
import pandas as pd
from pandas.tseries.offsets import DateOffset
import matplotlib.pyplot as plt
import statsmodels.api as sm

# --- PRELIMINARY SETUP ---

# Ensure that your index is in datetime format, sorted, and has monthly frequency.
df.index = pd.to_datetime(df.index)
df = df.sort_index()
df = df.asfreq('MS')  # 'MS' means Month Start; adjust if necessary

# --- MODEL FITTING ---

# Fit a SARIMAX model. Adjust (p,d,q) and seasonal_order as needed.
model = sm.tsa.statespace.SARIMAX(df['Sales'],
                                  order=(1,1,1),
                                  seasonal_order=(1,1,1,12))
results = model.fit()

# --- CREATING THE FUTURE DATE RANGE ---

# Create a list of 24 future dates. Adding DateOffset(months=x+1) starts forecasting
# from the month immediately following the last observation.
future_dates = [df.index[-1] + DateOffset(months=x+1) for x in range(24)]

# Create a new DataFrame with these dates and the same columns as df
future_df = pd.DataFrame(index=future_dates, columns=df.columns)

# --- COMBINING HISTORICAL DATA WITH FUTURE DATES ---

# Concatenate the original DataFrame with future_df (which has empty Sales values)
combined_df = pd.concat([df, future_df])

# --- FORECASTING ---

# Use the SARIMAX model's predict method.
# Note: We use the datetime index by specifying start and end as datetime values.
# Here, start is the first future date (which is at index len(df)) and end is the last index.
combined_df['forecast'] = results.predict(
    start=combined_df.index[len(df)], 
    end=combined_df.index[-1], 
    dynamic=True
)

# --- PLOTTING THE RESULTS ---

plt.figure(figsize=(12, 8))
plt.plot(combined_df['Sales'], label='Historical Sales', color='blue')
plt.plot(combined_df['forecast'], label='Forecast', color='orange')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title("Historical vs Forecasted Sales")
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

# 1. Ensure datetime index and monthly frequency
df.index = pd.to_datetime(df.index)
df = df.sort_index()
df = df.asfreq('MS')

# 2. Split into train and test (last 12 months as test)
train = df.iloc[:-12]
test = df.iloc[-12:]

# 3. Fit SARIMAX on train set
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(train['Sales'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
results = model.fit()

# 4. Predict for test period
forecast = results.predict(start=test.index[0], end=test.index[-1], dynamic=False)

# 5. Calculate performance metrics
mae = mean_absolute_error(test['Sales'], forecast)
rmse = np.sqrt(mean_squared_error(test['Sales'], forecast))
mape = mean_absolute_percentage_error(test['Sales'], forecast) * 100

# 6. Print metrics
print("📊 Forecasting Accuracy on Last 12 Months:")
print(f"MAE  = {mae:.2f}")
print(f"RMSE = {rmse:.2f}")
print(f"MAPE = {mape:.2f}%")

# 7. Optional: Plot actual vs forecast
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(train['Sales'], label='Train')
plt.plot(test['Sales'], label='Actual')
plt.plot(forecast, label='Forecast', color='orange')
plt.legend()
plt.title("Forecast Accuracy on Last 12 Months")
plt.grid(True)
plt.show()
```
